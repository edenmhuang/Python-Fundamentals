1. Write a Python function that determines the number of unique words in a text file, example, "The Complete Works of William Shakespeare" and counts how many times each word occurs. It should accept the path to a text file as the input argument and then print the total number of words in the file, the top 19 most frequently used words, and how many times each occurs. 

An example call to the function might look like this. Printing out a message with the total word count and a top 19 list. How you choose the format that output is really up to you. You should ignore the case when considering if two words are the same or different. For this exercise, you will say that words can contain letters, numbers, apostrophes, hyphens, and any other punctuation or white space that separates them. 


```python
import requests
import re
import collections
```

```python
def word(text):
    mycopy = open(text, 'r', errors='ignore')
    book = mycopy.read()
    all_words = re.findall(r"[0-9a-zA-Z-'\']+", book)
    cap = [w.upper() for w in all_words] 
    m = collections.Counter(cap)
    top19 = m.most_common(19)
    number_of_words = len(cap)
    print(f"Total Words: {number_of_words}\n")
    print ("Top 19 Words:")
    for i in top19:
        print(i[0] + " "*(9 - len(i[0])) + str(i[1]))
```

```python
word('shakespeare.txt')
```
