3) Write a function <b>find_next_point.</b> This function should return a link as well as the end position of the link. Your task is to modify the function to detect the case when no link is found. If there is no link tag, the function should return: None, 0

```python
def find_next_point(page):
    while True:
        if '<a href="' in page:
            start = page.find('<a href="') + len('<a href="')
            end = page.find('"', start)
            url = page[start:end]
            page = page[end:]
            return(url, end)
        else:
            return None, 0
```

```python
find_next_point(page)
```

5) Write a function <b>get_all_url.</b> This function must store all the URLs from the page to a list. Your function should return a list of all the URLs.

```python
def get_all_url(page):
    hist = []
    while '<a href="' in page:
        if page.find('<a href="') == -1: 
            break
        else:
            start = page.find('<a href="') + len('<a href="')
            end = page.find('"', start)
            url = page[start:end]
            page = page[end:]
            hist.append(url)
    print(hist)
    
get_all_url(page)
```
